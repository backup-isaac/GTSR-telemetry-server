```python
def checkTimes(start,end,merged):
    
    new_start = -1
    new_end = -1

    # Edge cases
    if start >= end:
        return "Invalid Times"
    
    if start < merged[0][0] or start > merged[-1][1]:
        new_start = start

    if end < merged[0][0] or end > merged[-1][1]:
        new_end = end
    
    # Get overall time range
    for pair in merged:
        
        # Check start if still in initial state
        if new_start == -1:
            if start < pair[0]:
                new_start = start
            elif start == pair[0] or start <= pair[1]:
                new_start = pair[1] + 1
        
        # Check end if still in initial state
        if new_end == -1:
            if end < pair[0]:
                new_end = end
            elif end == pair[0] or end <= pair[1]:
                new_end = pair[0] - 1
        
        
        # Check if new_start and new_end were found
        if new_start != -1 and new_end != -1:
            
            # Check if new_start and new_end were found in same pair
            if new_start > new_end:
                return "Times already merged"
            break
    
    # Remove any internal pairs from overall range
    pairs = []
    p1 = new_start
    p2 = new_end
    for pair in merged:
        
        if pair[1] < new_start:
            continue
        if pair[0] > new_end:
            pairs.append((p1,p2))
            break
        
        pairs.append((p1,pair[0]-1))
        p1 = pair[1]+1
    
    # Edge case
    if merged[-1][1] < new_end:
        pairs.append((p1,p2))
    return pairs
```

```python
# Merges added start/end time pairs into merged list
def addTimes(pairs,merged):
    
    # Initialize Variables
    new_list = []
    p1 = 0
    p2 = 0
    
    while True:
        
        if p1 == len(pairs) and p2 == len(merged):
            break
        
        if p1 == len(pairs):
            new_list.append(merged[p2])
            p2 += 1
            continue
        else:
            v1 = pairs[p1][0]
            
        if p2 == len(merged):
            new_list.append(pairs[p1])
            p1 += 1
            continue
        else:
            v2 = merged[p2][0]
        
        if v1 < v2:
            new_list.append(pairs[p1])
            p1 += 1
        else:
            new_list.append(merged[p2])
            p2 += 1
    
    return new_list
        
```

```python
# Merges time pairs together that are right next to each other
def reduceTimes(merged):
    
    new_list = [merged[0]]
    
    new_list_index = 0
    for n in range(1,len(merged)):
        
        # Get two pairs
        pair1 = new_list[new_list_index]
        pair2 = merged[n]
        
        # Check if they can be merged
        if pair1[1] + 1 == pair2[0]:
            new_pair = (pair1[0],pair2[1])
            new_list[new_list_index] = new_pair
        else:
            new_list.append(pair2)
            new_list_index += 1
    
    return new_list
```

```python
# Test above functions
def test(start,end,merged):
    # Simulate checking for previously merged times
    # Outputs time ranges that have not been merged yet
    to_merge = checkTimes(start,end,merged)
    print("Start/End Pair: ", '(' + str(start) + ',' + str(end) + ')')
    print("Previous Pairs: ", merged)
    print()
    print("Pairs to Merge: ", to_merge)

    if to_merge != "Times already merged":
        # Simulates adding new non-merged times into merged list
        added = addTimes(to_merge,merged)
        print("Merged Times:   ", added)

        # Simulates combining pairs in merged list that are next to each other
        reduced = reduceTimes(added)
        print("Reduced Times:  ", reduced)
```

```python
# Previously merged times
merged = [(10,20),(30,50),(80,180),(200,300)]

# Time range to merge
start = 5
end = 500

test(start,end,merged)
```
